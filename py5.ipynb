```python
class Animal:
    def run(self):
        print("Animal Run..!!")
        print("Color:", self.color)
        
    def eat(self):
        print("Animal Eat..!!")

# Instantiating the first object
dog = Animal()
dog.color = "White"
dog.run()
dog.eat()

# Instantiating the second object
cat = Animal()
cat.color = "Black"
cat.run()
cat.eat()
```

```python
class Employee:
    # Parameterized Constructor
    def __init__(self, eid, ename):
        self.eid = eid
        self.ename = ename
        print("Constructor is called.")
        
    def details(self):
        print("Eid:", self.eid)
        print("Ename:", self.ename)

# Instantiating objects with passing positional parameters
ajay = Employee(101, "Ajay")
ajay.details()

vijay = Employee(102, "Vijay")
vijay.details()
```

```python
class A:
    def __init__(self):
        print("Class A constructor called.")
        
    def get(self):
        self.x = 10
        print("This is get from class A")

# Class B inherits from Class A
class B(A):
    def __init__(self):
        super().__init__()  # Triggers parent class A constructor initialization
        print("Class B constructor called.")
        
    def put(self):
        print("This is put from class B")
        print("X:", self.x)

# Runtime object workflow
objb = B()
objb.get()  # Initializes self.x inside the shared instance scope
objb.put()  # Safely accesses the inherited self.x attribute
```
