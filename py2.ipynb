```python
#Program to find the prime number
num=7
flag=0
for i in range(2,num):
    if num%2==0:
        flag=1
        break
if flag==0: 
    print(num, "Is a Prime Number")
else:
    print("Is Not a Prime Number")
```

```python
#Fibonacci Series
fn=0
sn=1
for i in range(0,8):
    tn=fn+sn
    print(fn)
    fn=sn
    sn=tn
```

```python
n = 5 
fact = 1 
for i in range(1, n + 1): 
    fact *= i 
print("Factorial of 5 is", fact)
```

```python
for n in range(1, 50): 
    if n < 2:
        continue
    flag = 0 
    for i in range(2, n): 
        if n % i == 0:
            flag = 1 
            break 
    if flag == 0:
        print(n, end=" ")
```

```python
# Parameterless Function (Default/Void)
def add(): 
    a = 10 
    b = 5 
    c = a + b 
    print("Add:", c) 
add()
```

```python
# Parameterized Subtraction
def sub(a, b): 
    c = a - b 
    print("SUB:", c) 
sub(10, 5)
```

```python
# Return Expression Division
def div(a, b): 
    c = a / b 
    return c 
z = div(10, 2) 
print("Div:", z)
```

```python
# Standard Encapsulated Factorial Function
def factorial(n): 
    fact = 1 
    for i in range(1, n + 1): 
        fact = fact * i 
    return fact 
print("Factorial is:", factorial(5))
```

```python
# Program to calculate the area of circle, rectangle, and Square Using function and return value
def circle():  
    r = float(input("Enter Radius of the Circle: ")) 
    pi = 3.14  
    return pi * r * r 

def rect(): 
    l = float(input("Enter Length of the Rectangle: ")) 
    w = float(input("Enter Width of the Rectangle: ")) 
    return l * w 

def square(): 
    s = float(input("Enter Side of the Square: ")) 
    return s * s 

# Running Statements
print("Area of Circle:", circle())  
print("Area of Rectangle:", rect()) 
print("Area of Square:", square())  


# Fixed multiplication syntax errors from the handwriting logs
area_circle = lambda r, pi=3.14: pi * r * r 
area_rect = lambda l, w: l * w 
area_square = lambda s: s * s 

print("AREA OF CIRCLE:", area_circle(4))      
print("AREA OF RECTANGLE:", area_rect(4, 5))   
print("AREA OF SQUARE:", area_square(4))       
```
