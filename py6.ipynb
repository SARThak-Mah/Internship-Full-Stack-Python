```python
#Abstraction=>
from abc import ABC,abstractmethod
class Remote(ABC):

    @abstractmethod
    def ch_plus(self):
        pass

    @abstractmethod
    def ch_minus(self):
        pass

    def model(self):
        print("MOdel:3434")
        

class TV(Remote):
    def On(self):
        print("Tv is ON")
    
    def ch_plus(self):
        print("Ch plus active")

    def ch_minus(self):
        print("Ch minus active")
    
user=TV()
user.On()
user.ch_plus()
user.model()

#r=Remote()

```
