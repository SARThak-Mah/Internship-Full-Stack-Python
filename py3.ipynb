```python
# Creating List
my_list = [5, 10, 7, 8, 9, 4.5, "python"] 
print(type(my_list))  
print(len(my_list))

# Insert operation
my_list.insert(4, 20) 
my_list.append(100)    
print(my_list) 

# Pop operations
my_list.pop()      
my_list.pop(0) 
print(my_list)
my_list.remove(9)   
my_list.remove(7) 
print(my_list) 
```

```python
# Iterating over elements using sequential length range ranges
my_list = [6, 7, 8, 9, 4.5, "python"] 
for i in range(len(my_list)): 
    print(my_list[i]) 
```

```python
my_list=[5, 6, 7, 9, 4, 3]
for i in range(0,len(my_list)):
    if my_list[i]>5:
        print(my_list[i])
for j in my_list:
    if j>5:
        print(j)
```

```python
# Sequence Slicing Arrays [start:stop:step]
nums = [1, 2, 3, 4, 5, 6, 7, 9, 8] 
print(nums[-2])
print(nums[1:4])
print(nums[0:5])   
print(nums[:4])  
print(nums[1:])
print(nums[0:5:2])
print(nums[-5:-1])
```

```python
# Populating a list dynamically with loop updates
list = []
print("Enter length:")
n = int(input("Enter No:"))
print("Enter",n, " elements in lists:") 
for i in range(0, n): 
    j=int(input("Enter No:")) 
    list.append(j) 
print(list) 
```
