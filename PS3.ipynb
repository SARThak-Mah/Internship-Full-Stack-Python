```python
str = "Sarthak"
vowel_count = 0
for i in str:
    if i == 'a':
        vowel_count += 1
    elif i == 'e':
        vowel_count += 1
    elif i == 'i':
        vowel_count += 1
    elif i == 'o':
        vowel_count += 1
    elif i == 'u':
        vowel_count += 1
    else:
        pass
print(vowel_count)
```

```python
def fib(n):
    if (n == 0 or n == 1):
        return n
    return fib(n-2) + fib(n-1) # fib(5-2) + fib(5-1) = fib(3) + fib(4) = (fib(3-2) + fib(3-1)) + (fib(4-2) + fib(4-1))...(fib(n-2) + fib(n-1))
x = fib(5)
print(x)
```
