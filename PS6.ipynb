```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def get_info(self):
        print(f"Name: {self.name}\nAge: {self.age}")
        
p = Person("Sarthak", 20)
p.get_info()
# p = Person()
# p.get_info("Sarthak", 20)
# print(f"Person name: {p.name}, Person age: {p.age}")
```

```python
class Animal:
    def sound(self):
        print("Some sound")

class Dog(Animal):
    def sound(self):
        print("Bark!")

a = Animal()
a.sound()

d = Dog()
d.sound()
```
