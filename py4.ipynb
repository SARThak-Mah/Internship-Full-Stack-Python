```python
# Tuple 
tpl = (7, 5, 9, 3, 2, 6, 5)
print(type(tpl))      
print(tpl[2])         
print(tpl)            

for i in range(0, len(tpl)):
    print(tpl[i], end=" ")

print() 


for x in tpl:
    print(x, end=" ")
print()


print(tpl.count(5))   
print(tpl.index(7))   


print(tpl[1:5])       
print(tpl[1::3])      
print(tpl[:2])        
print(tpl[-4:-1])     


for i in tpl[::2]:
    print(i, end=" ") 
print()


tpl1 = (100,)
print(type(tpl1))     
```

```python
# Unpacking tuple
color = ("Black", "Blue", "Red", "Pink")

c1 = color[0]

c1, c2, c3, c4 = color
print(c1, c4)         

c1, c2, *c = color    
print(c1, c)         

tpl2 = (4, 6, 8)
lst = list(tpl2)      
lst[0] = 100          
tpl1 = tuple(lst)     
print(tpl1)           
```

```python
# Sets 
s = {4, 6, 7, 8, 9, 3, "Python", 100, 12, 15, 3, 4}
print(type(s))       
print(len(s))         

s.add(120)           
s.remove(7)         
print(s)

for x in s:
    print(x, end=" ")
print()

set_a = {4, 8, 6, 9, 45, 6}
set_b = {45, 78, 40, 23, 12, 4}

print(set_a.union(set_b))        
print(set_a.intersection(set_b))   
```

```python
# Dictionary 
p = {
    "Name": "Ajay",
    "Age": 19,
    "id": 101,
    "skill": ["python", "java"]
}

print(type(p))        
print(p["Name"])      

p["Age"] = 25         
p.update({"salary": 50000}) 
print(p)

p.pop("Name")         
p.popitem()           
print(p)

print(p.values())     
print(p.keys())     
print(p.items())     

for x in p:
    print(x)          

for x in p:
    print(p[x])       
```

```python
s = {
    "ajay": {
        "sub1": 98,
        "sub2": 78,
        "sub3": 88
    },
    "Pankaj": {
        "sub1": 78,
        "sub2": 91,
        "sub3": 82
    },
    "Monalisa": {
        "sub1": 72,
        "sub2": 82,
        "sub3": 79
    }
}

print(s["ajay"]["sub1"]) 

print("\n--- Iterating Nested Dictionaries ---")

for x in s:
    print(x)
    for y in s[x]:
        print(s[x][y], end=" ")
    print()
    
for x in s.values():
    print(x)
    for y in x.values():
        print(y, end=" ")
    print()
```
