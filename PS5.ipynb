```python
my_set = {1, 2, 3, 3, 4}
print(my_set) # my_set = {1, 2, 3, 3, 4}
my_set.add(5)
my_set.remove(2)
print(my_set)
is_4_present = 4 in my_set
print(is_4_present)

```

```python
a = {1, 2, 3}
b = {3, 4, 5}
un = a.union(b)
print(un)
insec = a.intersection(b)
print(insec)
diff = a.difference(b)
print(diff)
```
